```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from sklearn.cluster import KMeans

import kmeans
import clusval
import distance

import shapes

golden = (1+np.sqrt(5))/2
```

```python

x1 = shapes.cluster_shape(3, 4, 50, 20, shapes.ellipse)
x2 = shapes.cluster_shape(5, 7, 50, 20, shapes.ellipse)
x3 = shapes.cluster_shape(9, 11, 50, 20, shapes.ellipse)

fig = plt.figure(figsize=(18, 6))
ax = fig.add_subplot(131)
shapes.plot_cluster_shapes(x1, ax)
ax = fig.add_subplot(132)
shapes.plot_cluster_shapes(x2, ax)
ax = fig.add_subplot(133)
shapes.plot_cluster_shapes(x3, ax)
```

```python
X = np.concatenate([x1,x2,x3])

z1 = [0]*len(x1)
z2 = [1]*len(x2)
z3 = [2]*len(x3)

Z = np.concatenate([z1,z2,z3])

ix = range(len(x1)+len(x2)+len(x3))
np.random.shuffle(ix)

X = np.array([X[i] for i in ix])
Z = np.array([Z[i] for i in ix])
```

```python
# Shape Clustering with Procrustes
J, M = kmeans.kmeans(3, X, distance.procrustes)
print clusval.class_error(J, Z)
```

```python
# Shape Clustering with K-means/Euclidean
X2 = X.reshape(150, 40)
km = KMeans(3)
r = km.fit(X2)
J2 = r.labels_
print clusval.class_error(J2, Z)
```

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

import mnistshape as dshape
import distance as dist

import cPickle, gzip
```

```python
f = gzip.open('data/mnist.pkl.gz', 'rb')
train_set, valid_set, test_set = cPickle.load(f)
f.close()
images, labels = train_set
```

```python
n = 10
idx = np.random.choice(range(len(images)), size=n, replace=False)
ims = images[idx].reshape((n, 28, 28))
print idx
```

```python
fig = plt.figure(figsize=(15, 5))
i = 1
for X in ims:
    ax = fig.add_subplot(3, n, i)
    ax.imshow(X, cmap=plt.cm.gray)
    ax.axis('off')

    # get contours of X
    cs = dshape.get_contours(X, X[np.where( (X > 0.1) & (X < 0.6) )].mean())

    ax = fig.add_subplot(3, n, n+i)
    for contour in cs:
        R = np.array([[0, -1], [1, 0]])
        contour = contour.dot(R) + np.array([[0, 28]])
        ax.plot(contour[:,0], contour[:,1], 'o', ms=3)
    ax.axis('off')
    ax.set_xlim([0, 28])
    ax.set_ylim([0, 28])

    # pick only the largest contour and fix orientation
    c = sorted(cs, key=len, reverse=True)[0]
    R = np.array([[0, -1], [1, 0]])
    c = c.dot(R) + np.array([[0, 28]])

    # interpolate largest contour with n points
    x, y = dshape.interpolate(c, n=100, s=5)

    ax = fig.add_subplot(3, n, 2*n+i)
    ax.plot(x, y, '-')
    ax.set_xlim([0, 28])
    ax.set_ylim([0, 28])
    ax.axis('off')
    i += 1
```
